```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("/content/sample_data/train (1).csv")
```

```python
df
```

```python
df = df[["airline_sentiment","text"]]
df.head()
```

```python
df.airline_sentiment.value_counts()
```

```python
df.replace(to_replace="positive",value="1",inplace=True)
df.replace(to_replace="negative",value="-1",inplace=True)
df.replace(to_replace="neutral",value="0",inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['text'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[:5]
```

```python
df["clean_text"] = corpus
```

```python
df.head()
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
voc_size = 5000
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
# Creating model
x = embedded_docs
y = df.airline_sentiment
```

```python
x.shape
```

```python
y.shape
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Create KNN classifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
